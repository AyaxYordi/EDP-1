La solución de la ecuación 
$$\frac{\partial ^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2}=0$$
sujeta a las condiciones de frontera:
$$u(x,0)=u(x,1)=0, \;\;\; 0<x<1$$
Y la condición inicial:
$$u(0,y)=10y,\hspace{1cm} u_x(1,y)=-1\;\;\;0<y<1$$

Esta dada en términos de la serie de Fourier por la expresión:
$$u(x,y)=\sum_{n=1}^{\infty}\left[ \frac{20(-1)^{n+1}}{n\pi}cosh(n\pi x)+\left( -20(-1)^{n+1}senh(n\pi)-\frac{2[(-1)^{n+1}+1]}{n\pi} \right)\frac{senh(n\pi x)}{n\pi cosh(n\pi)} \right]sen(n\pi y)$$

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.style as style
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
from math import *
```

Para empezar, graficamos la condición inicial. $$u(0,y)=10y$$

```python
f = np.poly1d([10,0])
```

```python
x = np.linspace(0,1,300)
plt.plot(x,f(x))
plt.xlabel('$y$'); plt.ylabel('$g(y)$'); 
plt.title("Condición inicial $u_x(1,y)=10y$")
plt.grid()
plt.show()
```

Usamos una serie de Fourier para aproximar la condición inicial $u(0,y)=10y$

```python
def fourier(x,N): # N es el número de armónicos
    s = 0
    for n in range(1,N+1):
        s += ((-1)**(n+1)/n)*sin(n*pi*x)
    return (20/pi)*s
```

```python
x1=[]
y1=[]
for i in range(1001):
  x1.append(0.001*i)
  y1.append(fourier(0.001*i,20))

plt.plot(x,f(x),label='$g(y)$')
plt.plot(x1,y1,label='$g_N(y)$ con $N=20$')
plt.xlabel('$y$'); plt.ylabel('$g(y)$'); 
plt.title("Aproximación de la condición inicial")
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

Podemos visualizar la aproximación para diferentes cantidades de armónicos.

```python
def Fourier(lista,N):
  l=[]
  for t in range(len(lista)):
    l.append(fourier(lista[t],N))
  return l

for i in range(1,20,4):
    plt.plot(x1,Fourier(x1,i),label='$g_N(y)$ con $N=%d$' %i)

plt.xlabel('$y$'); plt.ylabel('$g(y)$'); 
plt.title("Aproximación de la condición inicial")
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

```python
# funcion con 120 armonicos
def u(x,y):
  s = 0
  for n in range(1,121):
    s +=(20/pi)*((-1)**(n+1)/n)*cosh(n*pi*x)*sin(n*pi*y)-(20/pi)*((-1)**(n+1)/n)*tanh(n*pi)*sinh(n*pi*x)*sin(n*pi*y)-(2/pi**2)*(((-1)**(n+1)+1)/(cosh(n*pi)*n**2))*sinh(n*pi*x)*sin(n*pi*y)
  return s

y2=[]
for n2 in range(len(x1)):
  y2.append(u(0,x1[n2]))
```

Cabe mencionar que la función que se escribio en la ventana de aqui arriba ^ en forma de codigo fue:
$$ u(x,y)=\frac{20}{\pi}\sum_{n=1}^{∞} \frac{(-1)^{n+1}}{n}cosh(n\pi x)sin(n\pi y)-\frac{20}{\pi}\sum_{n=1}^{\infty} \frac{(-1)^{n+1}}{n}tanh(n\pi)senh(n\pi x)sen(n\pi y)-\frac{2}{\pi ^2}\sum_{n=1}^{\infty} \frac{(-1)^{n+1}}{n^2cosh(n\pi)}senh(n\pi x)sen(n\pi y)$$

Ahora, para visualizar la solución completa. Primero, para el primer punto, $y=0$.

```python
plt.plot(x1,y2,label='$u(0,y)$')
plt.xlabel('$y$'); plt.ylabel('$g(y)$'); 
plt.title("Aproximación de la condición inicial")
plt.grid()
plt.legend(loc='upper left')
plt.show()
```

Luego, para diferentes valores de $x$, lo cual nos permite ver cómo la solución evoluciona.

```python
def u_n(x,lista):
  l=[]
  for n in range(len(lista)):
    l.append(u(x,lista[n]))
  return l

for t2 in range(0,10):
  v = round( 0.1*t2 , 2 )
  plt.plot(x1,u_n(v,x1), label='u(%r,y)'%v)

plt.xlabel('$y$'); plt.ylabel('$u(x,y)$'); 
plt.title("Evolución de la solución")
plt.grid()
plt.legend(loc='upper left')
plt.show()
```

Y podemos ver esto en una gráfica en 3D.

```python
X = np.linspace(0,1,100)
Y = np.linspace(0,1,100)

X,Y=np.meshgrid(X,Y)

# función u(x,y)
def Z(x,y):
  m = np.zeros(np.shape(x))
  for i in range(np.shape(x)[0]):
    for j in range(np.shape(x)[1]):
      m[i][j]=u(x[i][j],y[i][j])
  return m

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X,Y,Z(X,Y), rstride=1, cstride=1, cmap=cm.jet)
plt.xlabel('x')
plt.ylabel('y')
plt.title('$u(x,y)$')
plt.show()
```
